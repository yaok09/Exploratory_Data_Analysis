```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
plt.subplot(221)

df['gender'].value_counts().plot(kind='bar', title='Gender of students', figsize=(16,9))

plt.xticks(rotation=0)

plt.subplot(222)

df['race/ethnicity'].value_counts().plot(kind='bar', title='Race/ethnicity of students')

plt.xticks(rotation=0)

plt.subplot(223)

df['lunch'].value_counts().plot(kind='bar', title='Lunch status of students')

plt.xticks(rotation=0)

plt.subplot(224)

df['test preparation course'].value_counts().plot(kind='bar', title='Test preparation course')

plt.xticks(rotation=0)

plt.show()
```

```python
plt.subplot(221)

df['gender'].value_counts().plot(kind='pie', title='Gender of students', figsize=(16,9))

plt.xticks(rotation=0)

plt.subplot(222)

df['race/ethnicity'].value_counts().plot(kind='pie', title='Race/ethnicity of students')

plt.xticks(rotation=0)

plt.subplot(223)

df['lunch'].value_counts().plot(kind='pie', title='Lunch status of students')

plt.xticks(rotation=0)

plt.subplot(224)

df['test preparation course'].value_counts().plot(kind='pie', title='Test preparation course')

plt.xticks(rotation=0)

plt.show()
```

```python
df.boxplot()
plt.show()
```

```python
sns.displot(df['math score'], kde = True)
plt.show()
```

```python
sns.histplot(df['math score'])
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True,square=True)
plt.yticks(rotation=0)
plt.show()
```

```python
sns.relplot(x='math score', y='writing score', hue='gender', data=df)
plt.show()
```

```python
df.groupby('parental level of education')\
[['math score', 'reading score', 'writing score']].mean().T.plot(figsize=(12,8))
```

```python
df.groupby('test preparation course')\
[['math score', 'reading score', 'writing score']].mean().T.plot(kind='barh', figsize=(10,10))
```
